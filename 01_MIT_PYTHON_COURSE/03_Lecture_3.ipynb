### while loop

```python
where = input("You are in lost Forest> Go Left or Right?")
while where == 'right':
    where = input("You are in lost Forest> Go Left or Right?")
print("You go out of Lost Forest")
```

```python
where = input("Go left or right")
while where == 'right':
    where = input('Go left or right')
print('You got out')
```

```python
n = int(input("enter a number"))
while n > 0:
    print('X')
    n = n -1
```

```python
count = 0
where = input("Go left or right")
while where == 'right':
    count += 1
    if count > 2:
        print(":(")
    where = input('Go left or right')
print('You got out')
```

```python
n = 0
while n < 5:
    print(n)
    n += 1
```

```python
# Factorial

x = 4
i = 1
factorial = 1
while i <= x:
    factorial *= i
    i += 1
print(f'{x} factorial is {factorial}')
```

### for loops

```python
for n in range(5):
    print(n)
```

```python
for i in range(1, 4, 1):
    print(i)
```

```python
for j in range(1, 4,2):
    print(j * 2)
```

```python
for me in range(4, 0, -1):
    print("$"*me)
```

```python
sum = 0
for i in range(10):
    sum += i
print(sum)
```

```python
mysum = 0
start = 1
stop = 15
for i in range(start, stop):
    mysum += i
print(mysum)
```

```python
# for loop and range

x = 4
factorial = 1
for i in range(1, x+1, 1):
    factorial *= i
print(f'{x} factorial is {factorial}')
```
