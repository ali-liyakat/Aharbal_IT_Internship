```python
# Newton's Method
# Find roots of polynomial   f(g, x) =  g**3 - x = 0
# Successive approximation
# next_guess = guess - f(guess) / f`(guess)

x = int(input("Enter X to find cube root"))
g = int(input("Enter start guess"))
print("Current estimated cube = ", g**3)

next_g = g - ((g ** 3 - x) / (3 * g ** 2))
print("Next Guess = ", next_g)
```

```python
pset = 5
sleep = 8
if ( pset + sleep) > 24:
    print("Impossible")
elif (pset + sleep) >= 24:
    print("Full schedule")
else:
    leftover = abs(24 - pset - sleep)
    print(leftover, "h of free time")
print("End of Day")
```

```python
# Nested Barnching
x = int(input("enter number for x"))
y = int(input("enter number for y"))
if x == y:
    print("X and Y are equal")
    if y != 0:
        print("Therefore X/Y is", x/y)
elif x < y:
    print("X is smaller")
else:
    print("Y is amller")
print("Thanks")
```
