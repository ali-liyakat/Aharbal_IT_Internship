```python
# square Root

guess = 0
x = int(input("Enter a number"))
while guess**2 < x:
    guess += 1
if guess**2 == x:
    print("The square root of", x, "is", guess)
else:
    print(x, "is not a perfect square")
```

```python
# if the number is negative

guess = 0
neg_flag = False
x = int(input("Enter a number"))
if x < 0:
    neg_flag = True
while guess**2 < x:
    guess += 1
if guess**2 == x:
    print("The square root of", x, "is", guess)
else:
    print(x, "is not a perfect square")
    if neg_flag:
        print(x, "is negative")
```

```python
# Cube root with negative number

cube = int(input("Enter Number"))
for guess in range(abs(cube)+1):
    if guess**3 ==  abs(cube):
        if cube < 0:
            guess = -guess
        print("Cube Root of " + str(cube)+ " is "+ str(guess))
```

```python
# linear equation problem

for  alysa in range(11):
    for ben in range(11):
        for cindy in range(11):
            total = (alysa + ben + cindy == 10)
            two_less = (ben == alysa - 2)
            twice = (cindy ==  2 * alysa)
            if total and two_less and twice:
                print(f"Alysa sold {alysa} tickets")
                print(f"Ben sold {ben} tickets")
                print(f"Cindy sold {cindy} tickets")
```
